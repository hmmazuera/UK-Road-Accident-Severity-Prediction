Vehicle Data Preparation

This notebook prepares the official UK Department for Transport vehicle dataset.

The objectives are:

- Load the vehicle dataset.
- Keep only vehicles involved in collisions from 2010 onwards.
- Explore the dataset.
- Assess data quality.
- Handle missing values where necessary.
- Remove unnecessary variables.
- Save a clean dataset for the data integration stage.

Import Libraries

```python
import pandas as pd
import numpy as np
```

Load Dataset

Load the cleaned collision dataset and the official vehicle dataset.

Only vehicles associated with collisions from 2010 onwards are retained to ensure consistency across the project.

```python
collisions = pd.read_csv('/content/drive/MyDrive/PROYECT_1/collision_clean.csv')
```

```python
valid_collision_ids = set(collisions['collision_index'])
```

```python
df = pd.read_csv('/content/drive/MyDrive/PROYECT_1/dft-road-casualty-statistics-vehicle-1979-latest-published-year.csv')
```

```python
df = df[df['collision_index'].isin(valid_collision_ids)].copy()
```
