# Collision Data Preparation

This notebook prepares the official UK Department for Transport collision dataset.

Objectives:

- Load the collision dataset
- Explore its structure
- Assess data quality
- Handle missing values
- Correct data types where necessary
- Save a cleaned version for further processing

Import Libraries

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

Load Dataset

```python
DATA_PATH = Path('../data/raw')

df = pd.read_csv(DATA_PATH / '/content/drive/MyDrive/PROYECT_1/dft-road-casualty-statistics-collision-1979-latest-published-year.csv')
```

Filter Recent Records

The original dataset contains collision records from 1979 onwards.

To develop a model that better represents current road conditions, legislation and vehicle technologies, only collisions from **2010 onwards** are retained.

This still provides a large number of observations while improving the relevance of the training data.

```python
pd.set_option('display.max_columns', None)
```

```python
df = df[df['collision_year']>=2010].copy()
```

```python
df[
    [
        "collision_severity",
        "enhanced_severity_collision",
        "collision_injury_based",
        "collision_adjusted_severity_serious",
        "collision_adjusted_severity_slight"
    ]
].describe()
```

```python
df.drop(["enhanced_severity_collision",
        "collision_injury_based",
        "collision_adjusted_severity_serious",
        "collision_adjusted_severity_slight"], axis=1, inplace=True)
```

```python
df.drop('collision_ref_no',axis=1,inplace=True)
```

```python
df.head()
```

```python
historic_columns = [
    ("junction_detail_historic", "junction_detail"),
    ("pedestrian_crossing_human_control_historic", "pedestrian_crossing"),
    ("pedestrian_crossing_physical_facilities_historic", "pedestrian_crossing"),
    ("carriageway_hazards_historic", "carriageway_hazards")
]
```

```python
for old_col, new_col in historic_columns:
  print('-' * 50)
  print(f'{old_col}')
  print(df[old_col].value_counts(dropna=False).sort_index())
  print()
  print(f'{new_col}')
  print(df[new_col].value_counts(dropna=False).sort_index())
  print()
```

```python
df.drop(['junction_detail_historic', 'pedestrian_crossing_human_control_historic',
'pedestrian_crossing_physical_facilities_historic', 'carriageway_hazards_historic'],axis=1,inplace=True)
```

Save Clean Dataset

Export the cleaned dataset.

The resulting dataset will be used during the data integration stage together with the vehicle and casualty datasets.

```python
df.to_csv('/content/drive/MyDrive/PROYECT_1/collision_clean.csv', index=False)
```
